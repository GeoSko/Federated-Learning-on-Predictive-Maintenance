# Imports

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

# Load Datasets

```python
# Load Centralized datasets
path = "/content/drive/MyDrive/Thesis/Datasets/Pasteurizer_dataset/"
train_df = pd.read_csv(path+ "centralized/train_set.csv")
test_df = pd.read_csv(path+ "centralized/test_set.csv")

# Load Federated datasets
alice_df = pd.read_csv(path+ "federated/alice_set.csv")
bob_df = pd.read_csv(path+ "federated/bob_set.csv")
test_fed_df = pd.read_csv(path+ "federated/test_set.csv")
```

# Centralized

```python
train_df
```

We will chech the differences between the train and test set

```python
# Calulate max timeslot per id
train_max_timeslot = train_df.groupby(by="paster_id")["paster_timeslot"].max()
test_max_timeslot = test_df.groupby(by="paster_id")["paster_timeslot"].max()
```

```python
print(f"Train set has {train_max_timeslot.shape[0]} blocks of pasteurizetion with mean of max_timeslot: {train_max_timeslot.mean()} ")
print(f"Test set has {test_max_timeslot.shape[0]} blocks of pasteurizetion with mean of max_timeslot: {test_max_timeslot.mean()} ")
```

```python
# Similar distributions of mean of max_timeslot across the two datasets
```

```python
sns.displot(train_max_timeslot, kde=True)
```

```python
sns.displot(test_max_timeslot, kde=True)
```

```python
# Sometimes the paster_timeslot stays the same. This is because the machine is in mode PASTER_RUN=0
# PASTER_RUN column is not present in the dataset
```

```python
# We can see below that sometimes the timeslot stays the same but we keep sampling from the other sensors and the values change
train_df[train_df["paster_id"]==7].iloc[110:130].head(8)
```

```python
# So its useful to see also the distribution of the number of samples per id, in addition to max_timeslots we saw before
```

```python
train_num_sumples = train_df.groupby(by="paster_id").size()
test_num_sumples = test_df.groupby(by="paster_id").size()
```

```python
print(f"Train set has mean number of samples per id: {train_num_sumples.mean()} ")
print(f"Test set has mean number of samples per id: {test_num_sumples.mean()} ")
```

```python
# Similar distributions of max of samples across the two datasets
```

```python
sns.displot(train_num_sumples, kde=True)
```

```python
sns.displot(test_num_sumples, kde=True)
```
